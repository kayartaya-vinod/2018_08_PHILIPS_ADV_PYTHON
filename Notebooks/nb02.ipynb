# Querying and filtering data

```python
import pandas as pd
```

```python
chip = pd.read_table('./resources/chipotle.tsv')
```

```python

chip[chip.quantity > 5] 
```

```python
users = pd.read_table('./resources/u.user', sep='|')
```

While combining multiple conditions using & or | group the conditions using parantheses

```python
users[(users.occupation=='writer') & (users.gender=='F')]
```

```python
users.loc[:,:]  
```

```python
users.loc[0:5] 
```

```python
users.loc[0:3, :'gender']
```

```python
users.loc[:, ['gender','zip_code'] ] 
```

### iloc excludes the upper limit for column range

```python
users.iloc[:3, :3] 
```

```python
filter_condition = (users.gender=='M') & (users.occupation=='writer')
users.loc[filter_condition, :'age']
```
